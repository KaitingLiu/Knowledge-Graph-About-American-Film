```python
import spacy
from spacy import displacy
import re
```

```python
test=datasets[0]['text']
nlp = spacy.load("en_core_web_sm")
doc = nlp(test)
sent=[]
for sentence in doc.sents:
    sent.append(sentence)
```

```python
test_spo=[]
for s in sent:
    for j in range(18):
      sub=datasets[0]['spo_list'][j]['h']['name']
      obj=datasets[0]['spo_list'][j]['t']['name']
      rel=datasets[0]['spo_list'][j]['relation']
      s=str(s)
      if re.search(sub,s) and re.search(obj,s):
          test_spo.append({"text":s,"spo_list":[sub,rel,obj]})
          break
```
